### Korrelációs elemzés a BMI_mean és HDI összetevőivel

Fontosabb könyvtárak betöltése

```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Excel fájl beolvasása
df = pd.read_excel("HDI_osszevont_valt.xlsx")
```

```python
df
```

##### Numerikus oszlopok kiválasztása a korrelációs elemzéshez

```python
# Csak a numerikus oszlopokat tartalmazó DataFrame
numeric_df = df.select_dtypes(include=[np.number])
```

##### Korrelációs mátrix számítása

```python
# Korrelációs mátrix számítása
correlation_matrix = numeric_df.corr(method='pearson')

# Eredmények kiíratása
print(correlation_matrix)
```

##### Eredmények kiíratása excel file-ba

```python
# Korrelációs mátrix mentése Excel fájlba (opcionális)
correlation_matrix.to_excel('korrelacios_matrix6.xlsx')
```

#### Korrelációs eredménytábla

```python
numeric_df.corr()
```

##### Korrelációs hőtérkép

```python
import seaborn as sns
plt.figure(figsize = (7, 5))
sns.heatmap(numeric_df.corr(), annot = True)
```

##### Páronkénti korreláció

```python
#páronkénti kolleráció
sns.pairplot(df, height=2)
plt.show()
```
