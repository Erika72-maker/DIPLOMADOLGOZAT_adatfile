### A hazai szerzők és a saját számítású HDI indexek összehasonlító diagramja

Kapcsolódó melléklet: 2 sz. melléklet  és saját HDI számítás alapján szerkesztett excel file az elemzés alapja

#### Fontosabb könyvtárak betöltése

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Adatok betöltése
file_path = 'HDI_comparison_results6.xlsx'
data = pd.read_excel(file_path)
```

```python
# DataFrame létrehozása
hdi_comparison = pd.DataFrame(data)
```

```python
# Elemzendő oszlopok
columns_to_plot = ["Magyar_HD _Riport _ 1990", "Csite-Németh-becslés_1994", "Magyar_HD_Riport _1996",
                   "Magyar_HD_Riport _1999", "Husz_Ildikó_becslése_2000", "Smahó_Melinda_becslése_2001",
                   "Csite-Németh-becslés_2001", "Csite-Németh-becslés_2005", "HDI","HDI_modified", "HDI_weighted"]
```

```python
# Átlagok számítása az egyes HDI típusokra
means = hdi_comparison[columns_to_plot].mean()
```

#### Vizualizáció

```python
# Oszlopok közötti távolság beállítása
x = np.arange(len(columns_to_plot))
width = 0.6  # Oszlopok szélessége
```

```python
# Diagram létrehozása
plt.figure(figsize=(14, 7))
bars = plt.bar(x, means, color=plt.cm.tab10(np.arange(len(columns_to_plot))), edgecolor='black')

# Címkék beállítása
plt.xticks(x, columns_to_plot, rotation=45, ha="right", fontsize=10)
plt.ylabel('Átlagos HDI érték', fontsize=12)
plt.title('HDI verziók összehasonlítása – Csoportosított oszlopdiagram', fontsize=14)

# Értékek kiírása az oszlopok tetejére
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.005, f"{height:.3f}", ha='center', fontsize=10, fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
