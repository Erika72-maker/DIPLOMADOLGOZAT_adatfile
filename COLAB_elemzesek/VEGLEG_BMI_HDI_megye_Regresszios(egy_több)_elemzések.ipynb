HDI és HDI változatok megyei szintű regressziós elemzése

```python
import pandas as pd      
import statsmodels.api as sm
```

#### Kétváltozós lineáris regresszió:  HDI és BMI átlag

```python


# DataFrame létrehozása
df = pd.read_excel("HDI_jo.xlsx")

# Független változó (HDI) és célváltozó (BMI_mean)
X = df["HDI"]
y = df["BMI_mean"]

# Konstans hozzáadása a modellhez (intercept)
X = sm.add_constant(X)

# Lineáris regresszió illesztése
model = sm.OLS(y, X).fit()

# Eredmények kiíratása
coefficients = model.params
print("Regressziós együtthatók:")
print(coefficients)

# Modellszummázás
summary = model.summary()
summary
```

#### Kétváltozós lineáris regresszió: HDI_modified és BMI átlag

```python


# Független változó (HDI) és célváltozó (BMI_mean)
X = df["HDI_modified"]
y = df["BMI_mean"]

# Konstans hozzáadása a modellhez (intercept)
X = sm.add_constant(X)

# Lineáris regresszió illesztése
model = sm.OLS(y, X).fit()

# Eredmények kiíratása
coefficients = model.params
print("Regressziós együtthatók:")
print(coefficients)

# Modellszummázás
summary = model.summary()
summary
```

#### Kétváltozós lineáris regresszió: HDI_weighted és BMI átlag

```python
# Független változó (HDI) és célváltozó (BMI_mean)
X = df["HDI_weighted"]
y = df["BMI_mean"]

# Konstans hozzáadása a modellhez (intercept)
X = sm.add_constant(X)

# Lineáris regresszió illesztése
model = sm.OLS(y, X).fit()

# Eredmények kiíratása
coefficients = model.params
print("Regressziós együtthatók:")
print(coefficients)

# Modellszummázás
summary = model.summary()
summary
```

#### Scatterplotok

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
df = pd.read_excel("HDI_jo.xlsx")
# Scatterplot és regressziós egyenes mindhárom HDI változóra
hdi_columns = ["HDI", "HDI_modified", "HDI_weighted"]

plt.figure(figsize=(15, 5))
for i, hdi_type in enumerate(hdi_columns, 1):
    X = df[hdi_type].values.reshape(-1, 1)
    y = df["BMI_mean"].values
    
    # Lineáris regresszió illesztése
    reg = LinearRegression().fit(X, y)
    y_pred = reg.predict(X)
    
    # Korreláció számítása
    correlation, _ = pearsonr(df[hdi_type], df["BMI_mean"])
    
    # Scatterplot
    plt.subplot(1, 3, i)
    plt.scatter(df[hdi_type], df["BMI_mean"], alpha=0.7, color="blue", label="Adatok")
    plt.plot(df[hdi_type], y_pred, color="red", label="Lineáris regresszió")
    plt.title(f"{hdi_type} vs BMI_mean\nr = {correlation:.3f}")
    plt.xlabel(hdi_type)
    plt.ylabel("BMI_mean")
    plt.legend()
    plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

### Többváltozós modell

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Adatok beolvasása (a képen szereplő adatok alapján, manuálisan beírva)
df = pd.read_excel("HDI_jo.xlsx")
#df = pd.DataFrame(data)

# Többszörös lineáris regresszió
# Függő változó: BMI_mean
# Független változók: HDI, HDI_modified, HDI_weighted
model = ols('BMI_mean ~ HDI + HDI_modified + HDI_weighted', data=df).fit()

# Eredmények kiíratása
print(model.summary())

# Opcionális: Változók standardizálása (a jobb összehasonlíthatóság érdekében)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['HDI', 'HDI_modified', 'HDI_weighted', 'BMI_mean']] = scaler.fit_transform(df[['HDI', 'HDI_modified', 'HDI_weighted', 'BMI_mean']])

# Többszörös lineáris regresszió a standardizált változókkal
model_scaled = ols('BMI_mean ~ HDI + HDI_modified + HDI_weighted', data=df).fit()
print("\n--- Standardizált változókkal ---")
print(model_scaled.summary())
```

### Polinomiális regresszió

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

# Scatterplot és polinomiális regresszió mindhárom HDI változóra
hdi_columns = ["HDI", "HDI_modified", "HDI_weighted"]

plt.figure(figsize=(15, 5))
for i, hdi_type in enumerate(hdi_columns, 1):
    X = df[hdi_type].values.reshape(-1, 1)
    y = df["BMI_mean"].values
    
    # Polinomiális transzformáció (másodfokú)
    poly = PolynomialFeatures(degree=2)
    X_poly = poly.fit_transform(X)
    
    # Lineáris regresszió illesztése a polinomiálisan transzformált adatokra
    reg = LinearRegression().fit(X_poly, y)
    y_pred = reg.predict(X_poly)
    
    # Korreláció számítása
    correlation, _ = pearsonr(df[hdi_type], df["BMI_mean"])
    
    # Scatterplot
    plt.subplot(1, 3, i)
    plt.scatter(df[hdi_type], df["BMI_mean"], alpha=0.7, color="blue", label="Adatok")
    plt.plot(df[hdi_type], y_pred, color="red", label="Polinomiális regresszió")
    plt.title(f"{hdi_type} vs BMI_mean\nr = {correlation:.3f}")
    plt.xlabel(hdi_type)
    plt.ylabel("BMI_mean")
    plt.legend()
    plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

Polinomiális transzformáció: A másodfokú (kvadratikus) regressziót készíti el minden HDI-változóra.
Scatterplot és illesztés: A scatterplot kék pontokkal ábrázolja az adatokat, míg a piros görbe a polinomiális illesztést mutatja.
Korreláció számítása: Az ábra címében megjelenik az r érték (Pearson-korrelációs együttható).
